````python
x = x.reshape(-1, 1)

print(f'X shape = {x.shape}')
print(f'X = {x}')
````

Kode ini melakukan perubahan bentuk (reshape) pada array `x` yang sebelumnya berisi data dari kolom `'luas'`. Berikut adalah penjelasan baris demi baris:

1. **Merubah bentuk array `x`**:  
   - `x = x.reshape(-1, 1)`  
     Fungsi `reshape()` digunakan untuk mengubah dimensi array. Parameter `(-1, 1)` menginstruksikan NumPy untuk mengubah `x` menjadi array dua dimensi:
     - `-1` berarti jumlah baris akan disesuaikan secara otomatis berdasarkan jumlah elemen dalam array.
     - `1` berarti setiap baris akan memiliki 1 kolom.
     
     Sehingga, array `x` yang sebelumnya berbentuk satu dimensi seperti `[6, 8, 10, 14, 18]`, akan diubah menjadi dua dimensi dengan satu kolom dan lima baris:

     ```

     ```

2. **Mencetak bentuk array `x`**:  
   - `print(f'X shape = {x.shape}')`  
     Baris ini mencetak bentuk (shape) array `x` setelah diubah. Hasilnya akan menunjukkan bahwa `x` sekarang memiliki 5 baris dan 1 kolom, yaitu `(5, 1)`.

3. **Mencetak array `x`**:  
   - `print(f'X = {x}')`  
     Baris ini mencetak nilai dari array `x` setelah diubah bentuknya.

Jadi, array `x` kini berbentuk dua dimensi dengan lima baris dan satu kolom.

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
````

Kode ini menggunakan pustaka `scikit-learn` untuk membangun dan melatih sebuah model **regresi linier**. Berikut adalah penjelasan baris demi baris:

1. **Import `LinearRegression`**:  
   - `from sklearn.linear_model import LinearRegression`  
     Baris ini mengimpor kelas `LinearRegression` dari pustaka `sklearn.linear_model`, yang digunakan untuk melakukan analisis regresi linier.

2. **Membuat objek model**:  
   - `model = LinearRegression()`  
     Baris ini membuat objek `model` yang merupakan instansi dari kelas `LinearRegression`. Objek ini nantinya akan digunakan untuk melatih model regresi linier.

3. **Melatih model dengan data**:  
   - `model.fit(x, y)`  
     Fungsi `fit()` digunakan untuk melatih model regresi linier dengan data yang diberikan. Parameter `x` adalah array input (variabel independen, dalam hal ini luas tanah), dan `y` adalah array target (variabel dependen, dalam hal ini harga tanah).  
     Fungsi ini akan menemukan hubungan linier antara `x` dan `y` berdasarkan data yang ada, dan model akan mempelajari koefisien regresi yang paling sesuai.

Setelah kode ini dijalankan, model regresi linier akan "belajar" dari data dan siap untuk digunakan untuk prediksi atau evaluasi lebih lanjut, seperti menghitung koefisien regresi atau membuat prediksi harga tanah berdasarkan luas tanah.

````python
x_vis = np.array([0,25]).reshape(-1, 1)
y_vis = model.predict(x_vis)
````

Kode ini digunakan untuk memprediksi harga tanah berdasarkan dua nilai luas tanah (0 dan 25), dengan menggunakan model regresi linier yang telah dilatih sebelumnya. Berikut adalah penjelasan baris demi baris:

1. **Membuat array untuk nilai `x` yang akan diprediksi**:  
   - `x_vis = np.array([0, 25]).reshape(-1, 1)`  
     Baris ini membuat array `x_vis` yang berisi dua nilai, yaitu 0 dan 25. Nilai-nilai ini mewakili dua luas tanah yang ingin diprediksi harga tanahnya oleh model regresi linier.  
     Fungsi `reshape(-1, 1)` digunakan untuk mengubah array `x_vis` menjadi bentuk dua dimensi (dengan satu kolom), karena model regresi linier mengharapkan input dalam bentuk array dua dimensi.

2. **Melakukan prediksi menggunakan model**:  
   - `y_vis = model.predict(x_vis)`  
     Fungsi `predict()` digunakan untuk memprediksi nilai `y` (harga tanah) berdasarkan nilai `x_vis` (luas tanah). Model regresi linier yang telah dilatih sebelumnya akan memberikan prediksi harga tanah untuk masing-masing luas tanah yang ada dalam `x_vis`.  
     Hasil dari `model.predict(x_vis)` adalah array prediksi harga tanah untuk luas 0 dan 25.

Setelah kode ini dijalankan, `y_vis` akan berisi prediksi harga tanah untuk dua nilai luas tanah yang dimasukkan (0 dan 25).

# **Formula Linear Regresion**

````python
print(f'intercept = {model.intercept_}')
print(f'slope = {model.coef_}')
````

Kode ini digunakan untuk mencetak nilai **intercept** (titik potong pada sumbu Y) dan **slope** (kemiringan garis regresi) dari model regresi linier yang telah dilatih. Berikut adalah penjelasan baris demi baris:

1. **Mencetak intercept (titik potong)**:  
   - `print(f'intercept = {model.intercept_}')`  
     `model.intercept_` adalah atribut yang menyimpan nilai intercept dari model regresi linier. Intercept adalah nilai harga tanah ketika luas tanah (`x`) sama dengan 0. Ini adalah titik potong garis regresi dengan sumbu Y. Kode ini mencetak nilai intercept tersebut.

2. **Mencetak slope (kemiringan garis)**:  
   - `print(f'slope = {model.coef_}')`  
     `model.coef_` adalah atribut yang menyimpan nilai slope (koefisien regresi) dari model. Slope menunjukkan perubahan harga tanah (`y`) untuk setiap perubahan satu unit dalam luas tanah (`x`). Dalam regresi linier, ini menggambarkan kemiringan garis regresi. Kode ini mencetak nilai slope tersebut.

Setelah kode ini dijalankan, Anda akan melihat output yang berisi nilai **intercept** dan **slope** dari model regresi linier.

# **Training & Testing Dataset**

````python
x_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = np.array([7, 9, 13, 17.5, 18])

x_test = np.array([9, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
````

Kode ini mempersiapkan data untuk **pelatihan** (training) dan **pengujian** (testing) model regresi linier. Berikut adalah penjelasan baris demi baris:

### Data untuk Pelatihan (Training)

1. **Membuat data pelatihan `x_train` dan `y_train`**:
   - `x_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)`  
     Baris ini membuat array `x_train` yang berisi nilai-nilai luas tanah untuk data pelatihan (yaitu, `[6, 8, 10, 14, 18]`). Fungsi `reshape(-1, 1)` mengubah array menjadi dua dimensi (dengan satu kolom dan lima baris).
     Hasilnya:
     ```
     [[ 6]
      [ 8]
      [10]
      [14]
      [18]]
     ```

   - `y_train = np.array([7, 9, 13, 17.5, 18])`  
     Baris ini membuat array `y_train` yang berisi nilai-nilai harga tanah yang sesuai untuk data pelatihan (yaitu, `[7, 9, 13, 17.5, 18]`).

### Data untuk Pengujian (Testing)

2. **Membuat data pengujian `x_test` dan `y_test`**:
   - `x_test = np.array([9, 9, 11, 16, 12]).reshape(-1, 1)`  
     Baris ini membuat array `x_test` yang berisi nilai-nilai luas tanah untuk data pengujian (yaitu, `[9, 9, 11, 16, 12]`). Sama seperti sebelumnya, fungsi `reshape(-1, 1)` mengubah array menjadi dua dimensi (dengan satu kolom dan lima baris).
     Hasilnya:
     ```
     [[ 9]
      [ 9]
      [11]
      [16]
      [12]]
     ```

   - `y_test = np.array([11, 8.5, 15, 18, 11])`  
     Baris ini membuat array `y_test` yang berisi nilai-nilai harga tanah yang sesuai untuk data pengujian (yaitu, `[11, 8.5, 15, 18, 11]`).

### Tujuan
- **`x_train` dan `y_train`** digunakan untuk melatih (fit) model regresi linier.
- **`x_test` dan `y_test`** digunakan untuk menguji kemampuan prediksi model setelah dilatih.

Dengan data ini, Anda dapat melatih model menggunakan `x_train` dan `y_train`, kemudian menguji kinerja model menggunakan `x_test` dan `y_test`.

# **Training Simple Linear Regression**

````python
model = LinearRegression()
model.fit(x_train, y_train)
````

Kode ini digunakan untuk melatih model regresi linier menggunakan data pelatihan yang telah disiapkan. Berikut adalah penjelasan baris demi baris:

1. **Membuat objek model regresi linier**:
   - `model = LinearRegression()`  
     Baris ini membuat objek `model` yang merupakan instansi dari kelas `LinearRegression` dari pustaka `scikit-learn`. Objek ini akan digunakan untuk membuat model regresi linier.

2. **Melatih model dengan data pelatihan**:
   - `model.fit(x_train, y_train)`  
     Fungsi `fit()` digunakan untuk melatih model regresi linier dengan data pelatihan. Parameter `x_train` adalah array input (luas tanah), dan `y_train` adalah array target (harga tanah).  
     Fungsi ini akan menemukan hubungan linier antara `x_train` dan `y_train`, dan model akan mempelajari koefisien regresi yang paling sesuai.

Setelah kode ini dijalankan, model regresi linier akan "belajar" dari data pelatihan dan siap untuk digunakan untuk membuat prediksi atau evaluasi lebih lanjut.

# **Evaluasi Kinerja Model Regresi Linier **

````python
from sklearn.metrics import r2_score

y_pred = model.predict(x_test)
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
````

Kode ini digunakan untuk mengevaluasi kinerja model regresi linier dengan menghitung **koefisien determinasi (R-squared)**. Berikut adalah penjelasan baris demi baris:

1. **Import `r2_score` dari pustaka `sklearn.metrics`**:
   - `from sklearn.metrics import r2_score`  
     Baris ini mengimpor fungsi `r2_score` dari pustaka `sklearn.metrics`, yang digunakan untuk menghitung R-squared. R-squared adalah ukuran seberapa baik model regresi linier menjelaskan variabilitas dalam data target (`y_test`).

2. **Memprediksi nilai `y` untuk data pengujian**:
